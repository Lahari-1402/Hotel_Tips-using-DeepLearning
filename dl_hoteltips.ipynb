```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
dictionary1 = {'Female':0, 'Male':1}
df['sex'] = df['sex'].map(dictionary1)
df.head()
```

```python
df['smoker'].unique()
```

```python
dictionary2 = {'No':0, 'Yes':1}
df['smoker'] = df['smoker'].map(dictionary2)
df.head()
```

```python
df['day'].unique()
```

```python
dictionary3 = {'Sun':1, 'Sat':2,'Thur':3,'Fri':4}
df['day'] = df['day'].map(dictionary3)
df.head()
```

```python
df['time'].unique()
```

```python
dictionary4 = {'Dinner':0, 'Lunch':1}
df['time'] = df['time'].map(dictionary4)
df.head()
```

```python
x = df.drop('tip', axis=1)
y = df['tip']
```

```python
import tensorflow as tf
#step 1 : create neuron layers with neurons
layer1 = tf.keras.layers.Dense(units=4, input_dim = 6)
layer2 = tf.keras.layers.Dense(units=32)
layer3 = tf.keras.layers.Dense(units=32)
layer4 = tf.keras.layers.Dense(units=64)
layer5 = tf.keras.layers.Dense(units=64)
layer6 = tf.keras.layers.Dense(units=128)
layer7 = tf.keras.layers.Dense(units=64)
layer8 = tf.keras.layers.Dense(units=64)
layer9 = tf.keras.layers.Dense(units=1)
```

```python
#Step 2 : add all layers to model
model = tf.keras.models.Sequential()
model.add(layer1)
model.add(layer2)
model.add(layer3)
model.add(layer4)
model.add(layer5)
model.add(layer6)
model.add(layer7)
model.add(layer8)
model.add(layer9)
```

```python
#Step 3 : give optimizer, loss and metrics
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['mean_squared_error'])
```

```python
model.fit(x,y,epochs=50)
```

```python
import numpy as np
model.predict(np.array([[21.01,1,0,1,0,3]]))
```
